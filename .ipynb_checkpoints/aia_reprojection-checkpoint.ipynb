```python
def most_recent_map(_map):
    
    if _map.lower()=="aia":
        info = (a.Instrument("aia") & a.Wavelength(94*u.angstrom))
        look_back = {"days":2}
    elif _map.lower()=="stereo":
        info = (a.Source('STEREO_A') & a.Instrument("EUVI") & a.Wavelength(171*u.angstrom, 211*u.angstrom)) 
        look_back = {"weeks":4}
    else:
        print("Don't know what to do! Please set _map=\"aia\" or \"stereo\".")
        
    current_date = datetime.datetime.now().strftime("%Y-%m-%dT%H:%M:%S")
    
    past = datetime.datetime.now()-datetime.timedelta(**look_back)
    past_date = past.strftime("%Y-%m-%dT%H:%M:%S")

    startt = str(past_date)
    endt= str(current_date)

    result = Fido.search(a.Time(startt, endt), info)
    
    file_download = Fido.fetch(result[0, -1], site='ROB')
    aiamap = sunpy.map.Map(file_download[0])
    
    return(aiamap)
```

```python
def reprojection(obstime:str, center_x, center_y, layers):
    
    aiamap = most_recent_map(_map="aia")
    print("Got AIA map.")
    stereomap = most_recent_map(_map="stereo")
    print("Got STEREO-A map.")
    
    # for aia
    in_time = aiamap.date
    out_time = obstime
    out_frame = Helioprojective(observer='earth', obstime=out_time, rsun=aiamap.coordinate_frame.rsun)
    rot_frame = RotatedSunFrame(base=out_frame, rotated_time=in_time)

    out_shape = aiamap.data.shape
    out_center = SkyCoord(0*u.arcsec, 0*u.arcsec, frame=out_frame)
    header = sunpy.map.make_fitswcs_header(out_shape,
                                       out_center,
                                       scale=u.Quantity(aiamap.scale))
    
    out_wcs = WCS(header)
    out_wcs.coordinate_frame = rot_frame

    with transform_with_sun_center():
        arr, _ = reproject_interp(aiamap, out_wcs, out_shape)
    
    out_warp = sunpy.map.Map(arr, out_wcs)
    out_warp.plot_settings = aiamap.plot_settings

    fig = plt.figure(figsize=(12, 12))

    ax1 = fig.add_subplot(2, 2, 1, projection=aiamap)
    aiamap.plot( title=f'Original {in_time}')
    plt.colorbar()

    ax2 = fig.add_subplot(2, 2, 2, projection=out_warp)
    out_warp.plot(title=f"Reprojected to an Earth observer at {out_time}")
    plt.colorbar()
    
    nustar_fov_side = 12*60 # arcseconds

    for diff in layers:
        if sunpy.__version__=="3.1.0":
            # get boxes in pixels, newer Sunpy doesn't allow draw_rectangle here
            center_pix = [out_shape[1]/2, out_shape[0]/2]
            arc2pix = u.Quantity(aiamap.scale).value[0]
            plt.gca().add_patch(patches.Rectangle([center_pix[0]+(center_x-nustar_fov_side/2 - diff)/arc2pix,center_pix[1]+(center_y-nustar_fov_side/2 - diff)/arc2pix], (nustar_fov_side+2*diff)/arc2pix, (nustar_fov_side+2*diff)/arc2pix, facecolor="none", linewidth=2, edgecolor='red'))
        else:
            # Translate to bottom left corner of rectangle
            bottom_left = ((center_x - nustar_fov_side/2 - diff)*u.arcsec, (center_y - nustar_fov_side/2 - diff)*u.arcsec)
            rect_bl = SkyCoord(*bottom_left, frame=out_warp.coordinate_frame)
            out_warp.draw_rectangle(rect_bl, width=(nustar_fov_side+2*diff)*u.arcsec, height=(nustar_fov_side+2*diff)*u.arcsec, color='red')
    
    
    # for stereo
    in_time = stereomap.date
    out_time = obstime
    out_frame = Helioprojective(observer='earth', obstime=out_time, rsun=stereomap.coordinate_frame.rsun)
    rot_frame = RotatedSunFrame(base=out_frame, rotated_time=in_time)

    out_shape = stereomap.data.shape
    out_center = SkyCoord(0*u.arcsec, 0*u.arcsec, frame=out_frame)
    header = sunpy.map.make_fitswcs_header(out_shape,
                                       out_center,
                                       scale=u.Quantity(stereomap.scale))
    
    out_wcs = WCS(header)
    out_wcs.coordinate_frame = rot_frame

    with transform_with_sun_center():
        arr, _ = reproject_interp(aiamap, out_wcs, out_shape)
    
    out_warp = sunpy.map.Map(arr, out_wcs)
    out_warp.plot_settings = stereomap.plot_settings
    
    
    ax3 = fig.add_subplot(2, 2, 3, projection=stereomap)
    stereomap.plot( title=f'Original {in_time}')
    plt.colorbar()

    ax4 = fig.add_subplot(2, 2, 4, projection=out_warp)
    out_warp.plot(title=f"Reprojected to an Earth observer at {out_time}")
    plt.colorbar()

    for diff in layers:
        if sunpy.__version__=="3.1.0":
            # get boxes in pixels, newer Sunpy doesn't allow draw_rectangle here
            center_pix = [out_shape[1]/2, out_shape[0]/2]
            arc2pix = u.Quantity(stereomap.scale).value[0]
            plt.gca().add_patch(patches.Rectangle([center_pix[0]+(center_x-nustar_fov_side/2 - diff)/arc2pix,center_pix[1]+(center_y-nustar_fov_side/2 - diff)/arc2pix], (nustar_fov_side+2*diff)/arc2pix, (nustar_fov_side+2*diff)/arc2pix, facecolor="none", linewidth=2, edgecolor='red'))
        else:
            # Translate to bottom left corner of rectangle
            bottom_left = ((center_x - nustar_fov_side/2 - diff)*u.arcsec, (center_y - nustar_fov_side/2 - diff)*u.arcsec)
            rect_bl = SkyCoord(*bottom_left, frame=out_warp.coordinate_frame)
            out_warp.draw_rectangle(rect_bl, width=(nustar_fov_side+2*diff)*u.arcsec, height=(nustar_fov_side+2*diff)*u.arcsec, color='red')
    
    plt.show()
```
